```python
# Importing Libraries required for this to work properly.
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# Directories for specific files for exploration purposes.
gauss_blur = os.listdir('C:/Users/ANIRUDH/OneDrive/Desktop/Code Application/input/gaussian_blurred')
gauss_blur.sort()
sharp = os.listdir('C:/Users/ANIRUDH/OneDrive/Desktop/Code Application/input/sharp')
sharp.sort()
```

```python
# Comparision between the blur images and the sharp images
plt.figure(figsize=(15,12))

for i in range(3):
    blur_image = plt.imread(f"C:/Users/ANIRUDH/OneDrive/Desktop/Code Application/input/gaussian_blurred/{gauss_blur[i]}")
    blur_image = cv2.resize(blur_image, (224, 224))
    plt.subplot(1, 3, i+1)
    plt.imshow(blur_image)
plt.show()

plt.figure(figsize=(15,12))

for i in range(3):
    sharp_image = plt.imread(f"C:/Users/ANIRUDH/OneDrive/Desktop/Code Application/input/sharp/{sharp[i]}")
    sharp_image = cv2.resize(sharp_image, (224, 224))
    plt.subplot(1, 3, i+1)
    plt.imshow(sharp_image)
plt.show()
```

```python
# Checking if the images are related to each other.
x_blur = []
for i in range(len(gauss_blur)):
    x_blur.append(gauss_blur[i])

y_sharp = []
for i in range(len(sharp)):
    y_sharp.append(sharp[i])

print(x_blur[10])
print(y_sharp[10])
```
